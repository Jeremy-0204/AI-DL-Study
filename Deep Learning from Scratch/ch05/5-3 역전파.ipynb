### 5.3 역전파

이번 절에서는 '+', 'x' 등의 연산을 예로 들어 역전파의 구조를 설명한다. 

#### 5.3.1 덧셈 노드의 역전파

z = x + y라는 식의 각 편미분을 계산하면 1이 된다. 이를 계산 그래프로 확인해보면 역전파시에 결국 곱해지는 값은 1이 된다.  
이는 상류에서 전해진 미분에 1을 곱해서 하류로 흘리는 역할을 한다. 입력된 값을 그대로 다음 노드로 보내는 것이다.  
덧셈 노드 역전파는 입력 신호를 다음 노드로 출력할 뿐이다. 

#### 5.3.2 곱셈 노드의 역전파

z = xy라는 식의 편미분은 각각 y, x가 된다.  
이는 곧 상류의 값에 순전파 때의 입력 신호들을 서로 바꾼 값을 곱해 하류로 보내는 역할을 한다.  
순전파때 x였다면 역전파에서는 y, y였다면 x로 바꾼다는 이야기이다.  

입력시 x가 10, y가 5이고, 역전파로 상류에서 1.3이라는 값이 들어오면, x쪽으로는 1.3 * 5 = 6.5, y쪽으로는 1.3 * 10 = 13이 전달이 된다.  
곱셈의 역전파는 입력 신호를 바꾼 값을 곱한다. 덧셈 역전파는 그대로 흘려보내니 순방향의 입력신호가 필요 없는 반면, 곱셈의 경우는 순전파의 입력 신호를 변수에 저장을 해야 한다.

#### 5.3.3 사과 쇼핑의 예

사과의 가격, 사과의 개수, 소비세라는 세 변수 각각이 최종 금액에 어떻게 영향을 주는지를 풀어보자.  
이는 각각 사과 가격에 대한 지불 금액의 미분, 사과 개수에 대한 지불 금액의 미분, 소비세에 대한 지불 금액의 미분을 구하는 것에 해당될 것이다.  

곱셈 노드의 역전파는 입력 신호를 서로 바꿔 하류로 흘려보낸다.
