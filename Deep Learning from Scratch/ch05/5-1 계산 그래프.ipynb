### 5-1 계산 그래프

오차역전법을 위해 가중치의 기울기 계산을 설명할때는 수식으로 하는 법도 있지만 계산 그래프로 설명 가능하다.  

계산그래프란 계산 과정을 그래프로 나타낸 것이다. 그래프는 노드와 에지로 표현된다.  

#### 5.1.1 계산 그래프로 풀다

문제 1) 슈퍼에서 1개에 100원인 사과를 2개 샀다. 지불 금액을 구하라. 단, 소비세가 10% 부과된다.  

계산 그래프는 계산 과정을 노드와 화살표로 표현한다. 원 안에는 연산 내용을 적고 계산 결과는 화살표 위에 적어 결과가 왼쪽에서 오른쪽으로 전해지게 한다.  

사과 (100) -> x2 (200) -> x1.1 (220) -> 지불 금액  

여기서 곱셈을 하나의 연산으로 생각하고 두 입력에 대한 연산으로 처리할 수 있게 변경할 수도 있다.  
계산을 왼쪽에서 오른쪽으로 하는 것을 순전파라 하며, 반대로 오른쪽에서 왼쪽으로의 전파를 역전파라고 한다.  이는 미분 계산에 중요한 역할을 한다.  

#### 5.1.2 국소적 계산

계산 그래프의 특징은 국소적 계산을 전파함으로 최종 결과를 얻는다는데 있다.  
국소적 계산이란?? 자신과 직접 관계된 작은 범위라는 뜻이다.  
전체에서 어떤 일이 벌어지든 상관 없이 자신과 관계된 정보만으로 결과를 출력할 수 있다는 것이다.  

각 노드에서의 계산은 지금까지 어떤 복잡한 계산을 거쳐왔던 상관없이 당장 자신과 관련한 계산외에는 신경 쓸 게 없다.  
계산 그래프는 이처럼 국소적 계산에 집중하여 단순하지만 결과를 전달하여 복잡한 계산을 해내는 것이다.  

#### 5.1.3 왜 계산 그래프로 푸는가?

계산 그래프의 이점은 무엇인가??  
하나는 국소적 계산, 다른 하나는 중간 계산 결과를 모두 보관할 수 있다는 것이다.  
마지막으로 더 나아가 미분을 역전파를 통해 효율적으로 계산할 수 있는 점에 있다.  

만약 문제 1에서 사과 가격이 오르면 최종 금액에 어떤 영향을 끼치는지가 궁금하다고 해보자.  
이는 사과 가격에 대한 지불 금액의 미분을 구해야 한다. 사과 값이 아주 조금 올랐을 때 지불 금액이 얼마나 증가하느냐를 표시한다.
