### 4.1 데이터에서 학습한다!  

신경망의 특징은 데이터를 보고 학습한다는 것이다.  
학습이란 가중치 매개변수의 값을 데이터를 보고 자동으로 결정한다는 뜻이다.  

#### 4.1.1 데이터 주도 학습

기계학습은 데이터가 생명이다. 데이터에서 답과 패턴을 발견하고 이야기를 만든다.  
보통 사람이 문제를 해결할때는 경험과 직관을 단서로 패턴을 찾곤 한다.  
기계학습은 사람의 개입을 최소화하고 데이터로부터 패턴을 찾아낸다.  

이미지에서 특징(feature)을 추출하고 그 특징의 패턴을 기계학습 기술로 학습하는 방법이 있다.  
여기서 특징이란 입력 데이터에서 본질적인 데이터를 정확하게 추출할 수 있도록 설계된 변환기를 가리킨다.  
비전 분야에서는 SIFT, SURF, HOG 등의 특징을 많이 사용하여 이미지의 특징을 보통 벡터로 기술한다.  
그렇게 변환된 벡터를 가지고 지도 학습 방식의 대표 분류 기법인 SVM, KNN 등으로 학습할 수 있다.  

기계학습에서는 모아진 데이터로부터 규칙을 찾아내는 역할을 기계가 담당한다.  
하지만 신경망에서는 이미지를 있는 그대로 학습한다.  
이미지의 중요 특징까지도 기계가 스스로 학습하는 것이다.  
이를 보고 종단간 기계학습 end to end machine learning 이라고도 한다. 입력에서 출력을 사람의 개입 없이 얻는 것이다.  

#### 4.1.2 훈련 데이터와 시험 데이터

학습시는 훈련 데이터만 사용하여 최적의 매개변수를 찾는다. 그런 다음에 시험 데이터를 사용하여 실력을 평가한다.  
우리가 원하는 것은 범용적으로 사용 가능한 모델이다.  
범용 능력을 제대로 평가하기위해 훈련과 시험 데이터를 분리하는 것이다.   

범용 능력은 아직 보지 못한 데이터도 문제를 올바르게 풀어내는 능력이다.  
하나에만 지나치게 최적화되면 오버피팅이라 칭한다.
